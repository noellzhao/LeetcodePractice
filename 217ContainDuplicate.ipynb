# 217 Contains Duplicate

## Idea

1. Traverse each number
2. If any number was previously seen, then return True
3. If nothing returns true until the end, then return False

```python
from typing import List
class Solution:
    def containsDuplicate(self, nums: List[int]) -> bool:
        lookup = set()
        for num in nums:
            if num in lookup:
                return True
            else:
                lookup.add(num)
        return False
```

```python
# Manual testing
if __name__ == "__main__":
    sol = Solution()
    nums = [1,1,1,3,3,4,3,2,4,2]
    result = sol.containsDuplicate(nums)
    print(result)
```
