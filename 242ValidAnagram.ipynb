# Problem 242 Valid Anagram

## Method 1 - Hash
Idea:
1. Hash every character in input string in a dictionary and count their frequency.
2. Check with the other input string. If there is a match, then reduce the frequency by 1.
3. If there is any frequency goes down below 0, then there is a mismatch. Return False
4. If the program runs to the end, then return True

```python
class Solution:
    def isAnagram(self, s: str, t: str) -> bool:
        # hash every single character into a dictionary
        # check if any character is previously "seen"
        lookup = {}
        if len(s)!=len(t):
            return False
        else:
            for char in s:
                if char in lookup:
                    lookup[char]+=1
                else:
                    lookup[char] = 1
            for char in t:
                if char in lookup:
                    lookup[char] -= 1
                    if lookup[char]<0:
                        return False
                else:
                    return False
            return True
```

```python
# Manual testing
if __name__ == "__main__":
    sol = Solution()
    s = "anagram"
    t = "nagaram"
    result = sol.isAnagram(s, t)
    print(result)
```
