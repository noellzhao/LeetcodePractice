# 49 Group Anagrams

Idea:

1. for each word, sort their characters and use as key in dictionary
2. if there are more than one word that has the same key, append the original word in the dictionary
3. return the values of the dictionary

```python
from typing import List
class Solution:
    def groupAnagrams(self, strs: List[str]) -> List[List[str]]:
        # first, create an empty dictionary for output
        lookup = {}
        # then, try to merge anagram words by some indices
        for word in strs:
            key = "".join(sorted(word))
            if key in lookup:
                lookup[key].append(word)
            else:
                lookup[key] = [word]

        return list(lookup.values())
```
