# 347 Top K Frequent Elements

```python
nums = [1,2,1,2,1,2,3,1,3,2]
k = 2
```

```python
from typing import List
```

```python
class Solution:
    def topKFrequent(self, nums: List[int], k: int) -> List[int]:
        # 1. count frequency
        freq = {}
        for n in nums:
            freq[n] = freq.get(n, 0) + 1

        # 2. create bucket sort - index is the frequency
        buckets = [[] for _ in range(len(nums) + 1)]
        for n, c in freq.items():
            buckets[c].append(n)

        # 3. traverse from high frequency to low until k elements are collected
        res = []
        for c in range(len(nums), 0, -1):
            if not buckets[c]:
                continue
            for n in buckets[c]:
                res.append(n)
                if len(res) == k:
                    return res
```

```python
# manual testing
if __name__ == "__main__":
    sol = Solution()
    result = sol.topKFrequent(nums, k)
    print(result)
```
